```python
def add_word(d, word):
    if not word:
        return d
    if word[0] not in d:
        d[word[0]] = {}
    if len(word) == 1:
        d[word[0]]['.'] = {}
    d[word[0]] = add_word(d[word[0]], word[1:])
    return d

d = {}
d = add_word(d, 'act')
d = add_word(d, 'at')
d
```

```python
words = {}
with open('corncob_lowercase.txt', 'r') as fle:
    for line in fle:
        word = line.strip('\n')
        words = add_word(words, word)
```

```python
def lookup_word(d, word):
    if not word:
        return True
    if word[0] not in d:
        return False
    return lookup_word(d[word[0]], word[1:])

lookup_word(words, 'aah')
```

```python
da_set = set()

def find_internal_words(d, word, prefix):
    if not word:
        return None
    for i, c in enumerate(word):
        if c in d:
            if '.' in d[c]:
#                 print(prefix + c)
                da_set.add(prefix+c)
            find_internal_words(d[c], word[i+1:], prefix+c)

find_internal_words(words, 'bisexualerasure', '')
for i in sorted(da_set):
    print(i, end=' ')
```

```python
baseword = 'bisexualerasure'
def display_poem(words):
#     for char in baseword:
#         print(char, end=' ')
    print('')
    for w in words:
        i = 0
        for char in w:
            indexes = [k for k, ltr in enumerate(baseword) if ltr == char if k >= i]
            j = indexes[0]
            while i < len(baseword):
                if j == i:
                    print(char, end=' ')
                    i += 1
                    break
                else:
                    print(' ', end=' ')
                    i += 1
        print('\n')
    print('\n')

display_poem(['seals'])
```

```python
poem = ["bisexual erasure"]

for p in poem:
    display_poem(p.split(' '))
```
